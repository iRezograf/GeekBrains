```python
# молодые хуже спаслись.
tf['Age_cat'] = pd.cut(tf.Age,10)
sns.catplot(data = tf,hue = 'Survived', x = 'Age_cat', kind='count', saturation=0.5)  
plt.xticks(rotation=90)
```

```python
# График распределения возрастов по классам кают "совпадает" с графиком гибели
sns.catplot(data = tf,hue = 'Pclass', x = 'Age_cat', kind='count', saturation=0.5)  
plt.xticks(rotation=90)
```

```python
# два встречных графика: число погибших убывает с ростом стоимости, а количество спасенных растет.
# видимо, в связи с дискриминацией бедных или с нижних палуб былосложенее спастись.
tf['Fare_cat'] = pd.qcut(tf.Fare,10)
sns.catplot(data = tf,hue = 'Survived', x = 'Fare_cat', kind='count', saturation=0.5) 
plt.xticks(rotation=90)
```

```python
import numpy as np
```

```python
corr = tf.corr()
```

```python
# посмотреть корреляцию между возрастом и стоимостью
mask = np.triu(np.ones_like(corr, dtype=bool))

# Создаем полотно для отображения большого графика
f, ax = plt.subplots(figsize=(11, 9))

# Создаем цветовую политру
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Визуализируем данные кореляции 
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

### Итоги:

1. Количество выживших напрямую зависит от стоимости билета (класса) каюты.
    * в первом классе соотношение выживших мужчин и выживших женщин более уравновешенное, т.е. можно предположить, 
        что при спасении обращали внимание не только на пол, но и "класс" пасажира
2. Больше всего погибло молодых. Нужно отметить, что большее их число было заселено в 3-м классе.
    * возможно из трюма спасться сложнее.
3. Женщин спаслось гораздо больше, чем мужчин: и абсолютно, и в процентном отношении.
